# Objetivo

Vamos criar uma base de dados fictícia para prever o crescimento das vendas com base no orçamento de marketing. Depois, implementaremos um modelo de regressão linear para analisar essa relação.

**Base de Dados Fictícia**

Vamos supor que temos os dados do orçamento de marketing (em milhares de reais) e as vendas (em milhares de unidades) para cada mês.

# Código em Python

*  Gerando a base de dados




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Definindo os parâmetros da base de dados
np.random.seed(42)  # Para reprodutibilidade
n_amostras = 100  # Número de amostras
orcamento_marketing = np.random.rand(n_amostras, 1) * 100  # Orçamento de marketing entre 0 e 100 mil reais
true_slope = 0.8  # Aumento esperado nas vendas por mil reais de orçamento de marketing
true_intercept = 30  # Vendas base sem investimento em marketing
noise = np.random.randn(n_amostras, 1) * 5  # Ruído aleatório

# Gerando as vendas
vendas = true_slope * orcamento_marketing + true_intercept + noise

# Criando um DataFrame para armazenar os dados
dados = pd.DataFrame(data={'Orcamento_Marketing': orcamento_marketing.flatten(), 'Vendas': vendas.flatten()})

# Exibindo os primeiros registros da base de dados
print(dados.head())

# Plotando os dados
plt.scatter(orcamento_marketing, vendas, color='blue')
plt.title('Relação entre Orçamento de Marketing e Vendas')
plt.xlabel('Orçamento de Marketing (mil reais)')
plt.ylabel('Vendas (mil unidades)')
plt.show()
```



*   Aplicando o modelo de regressão linear:

```python
from sklearn.linear_model import LinearRegression

# Separando as variáveis independentes e dependentes
X = dados[['Orcamento_Marketing']]
Y = dados['Vendas']

# Criando e ajustando o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(X, Y)

# Obtendo os coeficientes do modelo
slope = modelo.coef_[0]
intercept = modelo.intercept_
print(f"Inclinação (Slope): {slope}, Intercepto (Intercept): {intercept}")

# Fazendo previsões
vendas_previstas = modelo.predict(X)

# Plotando a reta de regressão
plt.scatter(X, Y, color='blue', label='Dados')
plt.plot(X, vendas_previstas, color='red', label='Reta de Regressão')
plt.title('Regressão Linear')
plt.xlabel('Orçamento de Marketing (mil reais)')
plt.ylabel('Vendas (mil unidades)')
plt.legend()
plt.show()
```
